# Time Series Line Plot

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
years = range(2010, 2021)
countries = ['Country A', 'Country B', 'Country C']
data = {'Year': [], 'Country': [], 'Happiness Score': []}

for year in years:
    for country in countries:
        data['Year'].append(year)
        data['Country'].append(country)
        data['Happiness Score'].append(np.random.uniform(0, 10))

df_time_series = pd.DataFrame(data)
plt.figure(figsize=(12, 6))
sns.lineplot(data=df_time_series, x='Year', y='Happiness Score', hue='Country')
plt.title('Happiness Score Over Time')
plt.ylabel('Happiness Score')
plt.xlabel('Year')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```

Time Series Line Plot:
This line plot shows the trend of happiness scores over time for each country.
Observations:
Some countries exhibit a relatively stable trend in happiness scores over the years.
Others may show fluctuations or a noticeable upward or downward trend.
Understanding these trends can provide insights into the factors influencing happiness in different countries over time.
